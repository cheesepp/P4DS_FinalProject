## EDA

### Mức độ sẵn sàng thảo luận với người giám sát
Đầu tiên ta sẽ nhóm người trả lời theo quy mô công ty

```python
survey_df.groupby('no_employees').value_counts()
```

```python
# Tạo bảng tần số và tỷ lệ phần trăm
freq = survey_df.groupby('no_employees').agg(lambda x: x.value_counts())
percent = survey_df.groupby('no_employees').agg(lambda x: x.value_counts(normalize=True) * 100)
```
